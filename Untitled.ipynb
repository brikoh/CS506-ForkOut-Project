```python
import json
import re
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import scipy.sparse.linalg
import matplotlib.pyplot as plt
import numpy as np
import pickle
from operator import itemgetter
from sklearn.decomposition import TruncatedSVD
from sklearn.cluster import KMeans
import scipy.cluster.hierarchy as hr
from sklearn.preprocessing import StandardScaler
from sklearn import mixture
import seaborn as sns
import sklearn.metrics as metrics
```

```python
def hasNumbers(i):
    return any(char.isdigit() for char in i)

def correctItems(a,m):
    a = a.strip()
    if (len(a.split(' ')) < 10 and a!=''):
        a = re.sub("[^a-zA-Z\s:]", "", a).title()
        m.append(a)
```

```python
f2 = open("menuFinal copy2.json").read()
menuRaw = json.loads(f2)
menuGet = 0
menus = []
for key,List in menuRaw.items():
    if (len(List)!=0):
        menuGet+=1
        menuItems = []
        menuItemsStr = ''
        menuInfo = {}
        for i in range(len(List)):
            for items in range(len(List[i])):
                if (re.search('[|]',List[i][items])):
                    for x in List[i][items].split('|'):
                        correctItems(x,menuItems)
                elif (re.search('[,]',List[i][items])):
                    for g in List[i][items].split(','):
                        correctItems(g,menuItems)
                elif (re.search('[a-zA-Z]', List[i][items])):
                    correctItems(List[i][items],menuItems)
        menuInfo['Website'] = key            
        menuInfo['Items'] = menuItems
        menuInfo['ItemsCount'] = len(menuItems)
        menus.append(menuInfo)
```

```python
for i in range(len(menus)):
    ItemsString = ''
    for items in menus[i]['Items']:
        ItemsString+=items + ' '
    menus[i]['ItemsString'] = ItemsString
```

```python
# k-means++
error = np.zeros(60)
for k in range(1,60):
    vectorsk = u[:,:k] @ np.diag(s[:k])
    kmeans = KMeans(init='k-means++', n_clusters=k)
    kmeans.fit_predict(vectorsk)
    error[k] = kmeans.inertia_
```

```python
plt.plot(range(1,len(error)),error[1:])
plt.xlabel('Number of clusters')
dummy = plt.ylabel('Error')

fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 10
fig_size[1] = 10
plt.rcParams["figure.figsize"] = fig_size

plt.show()
```

```python
max = len(u)
for k in range(1,20):
    vectorsk = u[:,:k] @ np.diag(s[:k])
    kmeans = KMeans(n_clusters=20, init='k-means++', max_iter=100, n_init=10, random_state=0)
    kmeans.fit_predict(vectorsk)
    labelsk = kmeans.labels_
    centroidsk = kmeans.cluster_centers_
    errork = kmeans.inertia_
```

```python
print("The total error of the clustering is: {}.".format(errork))
print('\nCluster labels:')
print(labelsk)
print('\nCluster centroids:')
print(centroidsk)
```
