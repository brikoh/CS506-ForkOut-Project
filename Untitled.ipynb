## ForkOut Final Project Data Analysis

```python
import json
import re
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import scipy.sparse.linalg
import matplotlib.pyplot as plt
import numpy as np
```

```python
def hasNumbers(i):
    return any(char.isdigit() for char in i)

def correctItems(a,m):
    a = a.strip()
    if (len(a.split(' ')) < 10 and a!=''):
        a = re.sub("[^a-zA-Z\s:]", "", a).title()
        m.append(a)
```

```python
f2 = open("menuFinal copy2.json").read()
menuRaw = json.loads(f2)
menuGet = 0
menus = []
for key,List in menuRaw.items():
    if (len(List)!=0):
        menuGet+=1
        menuItems = []
        menuItemsStr = ''
        menuInfo = {}
        for i in range(len(List)):
            for items in range(len(List[i])):
                if (re.search('[|]',List[i][items])):
                    for x in List[i][items].split('|'):
                        correctItems(x,menuItems)
                elif (re.search('[,]',List[i][items])):
                    for g in List[i][items].split(','):
                        correctItems(g,menuItems)
                elif (re.search('[a-zA-Z]', List[i][items])):
                    correctItems(List[i][items],menuItems)
        menuInfo['Website'] = key            
        menuInfo['Items'] = menuItems
        menuInfo['ItemsCount'] = len(menuItems)
        menus.append(menuInfo)
```

```python
for i in range(len(menus)):
    ItemsString = ''
    for items in menus[i]['Items']:
        ItemsString+=items + ' '
    menus[i]['ItemsString'] = ItemsString
```

```python
for x in range(2):
    print(menus[x]['Website'], '\n',menus[x]['Items'], '\n',menus[x]['ItemsString'], '\n',menus[x]['ItemsCount'], '\n')
```

```python
df_menus = pd.DataFrame(menus, columns = ['Website', 'Items', 'ItemsString', 'ItemsCount'])
df_menus
```

```python
vectorizer = TfidfVectorizer(stop_words='english', min_df=5,max_df=0.8)

dtm = vectorizer.fit_transform(df_menus['ItemsString'])
```

```python
print(dtm.shape)
```

```python
terms = vectorizer.get_feature_names()
print(terms)
```

```python
u,s,vt = scipy.sparse.linalg.svds(dtm)
```
